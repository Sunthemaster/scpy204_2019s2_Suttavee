```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read
data = pd.read_csv("COVID-19-geographic-disbtribution-worldwide-2020-04-12.csv")
data.head()
```

```python
#Thaialnd
thcase=0
thdeath=0
#China
cncase=0
cndeath=0
#USA
uscase=0
usdeath=0
#Saudi Arabia
sacase=0
sadeath=0
#Russia
ruscase=0
rusdeath=0
#Vietnam
vncase=0
vndeath=0
#South Korea
skcase=0
skdeath=0
#Italy
itcase=0
itdeath=0
#Israel
isrcase=0
isrdeath=0
#France
frcase=0
frdeath=0
for row in range(10332):
    #print (row)
    if data.at[row,'geoId'] == "TH" :
        thcase=thcase+data.at[row,'cases']
        thdeath=thdeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "CHN" :
        cncase=cncase+data.at[row,'cases']
        cndeath=cndeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "USA" :
        uscase=uscase+data.at[row,'cases']
        usdeath=usdeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "SAU" :
        sacase=sacase+data.at[row,'cases']
        sadeath=sadeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "RUS" :
        ruscase=ruscase+data.at[row,'cases']
        rusdeath=rusdeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "VNM" :
        vncase=vncase+data.at[row,'cases']
        vndeath=vndeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "KOR" :
        skcase=skcase+data.at[row,'cases']
        skdeath=skdeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "ITA" :
        itcase=itcase+data.at[row,'cases']
        itdeath=itdeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "ISR" :
        isrcase=isrcase+data.at[row,'cases']
        isrdeath=isrdeath+data.at[row,'deaths']
    elif data.at[row,'countryterritoryCode'] == "FRA" :
        frcase=frcase+data.at[row,'cases']
        frdeath=frdeath+data.at[row,'deaths']
```

```python
dict_case={'Thailand':thcase,'China':cncase,'USA':uscase,'Saudi Arabia':sacase,'Russia':ruscase
           ,'Vietnam':vncase,'South Korea':skcase,'Italy':itcase,'Israel':isrcase,'France':frcase}
case=pd.Series(dict_case)
dict_death={'Thailand':thdeath,'China':cndeath,'USA':usdeath,'Saudi Arabia':sadeath,'Russia':rusdeath
            ,'Vietnam':vndeath,'South Korea':skdeath,'Italy':itdeath,'Israel':isrdeath,'France':frdeath}
death=pd.Series(dict_death)
dict_population={'Thailand':69428524,'China':1392730000,'USA':327167434,'Saudi Arabia':33699947,'Russia':144478050
            ,'Vietnam':95540395,'South Korea':51635256,'Italy':60431283,'Israel':8883800,'France':66987244}
population=pd.Series(dict_population)
dfdata=pd.DataFrame({'cases':case,'deaths':death,'population':population,'%infectperpopulation':(case/population)*100
                     ,'%deathpercase':(death/case)*100})
dfdata
```

```python
#make case list in each cotry
#Thaialnd
thcase=0
thday=[]
#China
cncase=0
cnday=[]
#USA
uscase=0
usday=[]
#Saudi Arabia
sacase=0
saday=[]
#Russia
ruscase=0
rusday=[]
#Vietnam
vncase=0
vnday=[]
#South Korea
skcase=0
skday=[]
#Italy
itcase=0
itday=[]
#Israel
isrcase=0
isrday=[]
#France
frcase=0
frday=[]
for row in range(10332):
    row = 10331-row
    if data.at[row,'geoId'] == "TH" :
        thcase=thcase+data.at[row,'cases']
        thday.append(thcase)
    elif data.at[row,'countryterritoryCode'] == "CHN" :
        cncase=cncase+data.at[row,'cases']
        cnday.append(cncase)
    elif data.at[row,'countryterritoryCode'] == "USA" :
        uscase=uscase+data.at[row,'cases']
        usday.append(uscase)
    elif data.at[row,'countryterritoryCode'] == "SAU" :
        sacase=sacase+data.at[row,'cases']
        saday.append(sacase)
    elif data.at[row,'countryterritoryCode'] == "RUS" :
        ruscase=ruscase+data.at[row,'cases']
        rusday.append(ruscase)
    elif data.at[row,'countryterritoryCode'] == "VNM" :
        vncase=vncase+data.at[row,'cases']
        vnday.append(vncase)
    elif data.at[row,'countryterritoryCode'] == "KOR" :
        skcase=skcase+data.at[row,'cases']
        skday.append(skcase)
    elif data.at[row,'countryterritoryCode'] == "ITA" :
        itcase=itcase+data.at[row,'cases']
        itday.append(itcase)
    elif data.at[row,'countryterritoryCode'] == "ISR" :
        isrcase=isrcase+data.at[row,'cases']
        isrday.append(isrcase)
    elif data.at[row,'countryterritoryCode'] == "FRA" :
        frcase=frcase+data.at[row,'cases']
        frday.append(frcase)
```

```python
#plot case and day 
fig=plt.figure()
plt.plot(np.arange(len(thday))+1,thday,label='Thailand')
plt.plot(np.arange(len(cnday))+1,cnday,label='China')
plt.plot(np.arange(len(usday))+1,usday,label='USA')
plt.plot(np.arange(len(saday))+1,saday,label='Saudi Arabia')
plt.plot(np.arange(len(rusday))+1,rusday,label='Russia')
plt.plot(np.arange(len(vnday))+1,vnday,label='Vietnam')
plt.plot(np.arange(len(skday))+1,skday,label='South Korea')
plt.plot(np.arange(len(itday))+1,itday,label='Italy')
plt.plot(np.arange(len(isrday))+1,isrday,label='Israel')
plt.plot(np.arange(len(frday))+1,frday,label='France')

plt.yscale('log')
plt.legend(bbox_to_anchor=(1.02, 0.7),loc='upper left')
plt.title("Graph relation between number of day and Covid-19 cases in 10 countries")
plt.xlabel("Number of days (day)")
plt.ylabel("Covid-19 cases (cases)")
plt.xlim(0)
```

```python
fig=plt.figure()
color=["tab:blue","tab:orange","tab:green","tab:red","tab:purple"
       ,"tab:brown","tab:pink","tab:gray","tab:olive","tab:cyan"]
name=list(dfdata.index[:])
for row in range(10):
    plt.scatter(dfdata.iloc[row]['%infectperpopulation'],dfdata.iloc[row]['%deathpercase'],c=color[row],s=dfdata.iloc[row]['population']/500000,label=name[row],alpha=0.6)
plt.legend(bbox_to_anchor=(1.02, 0.8),loc='upper left')
plt.title("Graph relation between %infectperpopulation and %deathpercase 10 countries")
plt.xlabel("infectperpopulation (%)")
plt.ylabel("deathpercase (%)")
```
