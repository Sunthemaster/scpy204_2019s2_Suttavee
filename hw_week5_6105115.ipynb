```python
import matplotlib.pyplot as plt
from astroquery.sdss import SDSS
```

```python
query = "select a.objID, a.petroMag_g, b.z, a.petroMagErr_g, b.zErr	from PhotoObjAll a 	join SpecObjAll b on a.objID=b.bestobjID	where b.zWarning=0 and b.z between 0.3 and 0.4 and a.petroMag_g > -4000 and a.petroMagErr_g < 1"
res = SDSS.query_sql(query)
res
```

```python
query = "select a.objID, a.petroMag_g, b.z, a.petroMagErr_g	from PhotoObjAll a 	join SpecObjAll b on a.objID=b.bestobjID	where b.zWarning=0 and b.z between 0.3 and 0.4 and a.petroMag_g > -4000"
res = SDSS.query_sql(query)
fig=plt.figure(dpi=300)
#plt.scatter(list(res[2][:]),list(res[1][:]),s=1)
plt.scatter(list(res[2][:]), list(res[1][:]),s=1)
plt.xlim(0.3,0.4)
plt.title("Graph relation between z and petroMag_g")
plt.xlabel("z")
plt.ylabel("petroMag_g (mag)")
```

```python
query = "select a.objID, a.petroMag_g, b.z, a.petroMagErr_g, b.zErr	from PhotoObjAll a 	join SpecObjAll b on a.objID=b.bestobjID	where b.zWarning=0 and b.z between 0.3 and 0.4 and a.petroMag_g > -4000 and a.petroMagErr_g < 5 and b.zErr < 0.05"
res = SDSS.query_sql(query)
fig=plt.figure(dpi=300)
plt.errorbar(list(res[2][:]), list(res[1][:]), yerr=list(res[3][:]),xerr=list(res[4][:]),linestyle="None")
plt.xlim(0.3,0.4)
plt.title("Graph relation between z and petroMag_g")
plt.xlabel("z")
plt.ylabel("petroMag_g (mag)")
```
